```
explanatory_variables =["var1","Income"]
target_variable= ["y"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data[explanatory_variables], data[target_variable], test_size=0.3,random_state=42)
summary_df = pd.DataFrame({
  " Train (Counts)" :X_train["Income"].value_counts(),
  "Test (Counts)" :X_test["Income"].value_counts(),
  " Train (Percent)" :X_train["Income"].value_counts()*100/len(X_train),
  "Test (Percent)" :X_test["Income"].value_counts()*100/len(X_test)  
}).sort_index()
print(summary_df)
```
